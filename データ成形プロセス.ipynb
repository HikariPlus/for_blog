- feature.tsvとfeature\_~.csvをBoW表現にする

```python
data_feature = pd.read_csv('feature.tsv',delimiter='\t')
data_feature2017 = pd.read_csv('feature_2017autumn.csv')
data_preference = pd.read_csv('preference.tsv',delimiter='\t')
```

```python
data_allfeature=data_feature
data_allfeature.append(data_feature2017)
```

```python
supvis = np.array(pd.unique(data_allfeature['監督']),dtype=np.string_)
story = np.array(pd.unique(data_allfeature['シリーズ構成']),dtype=np.string_)
company = np.array(pd.unique(data_allfeature['制作会社']),dtype=np.string_)
```

```python
Indexes = np.array(Indexes,dtype=np.string_)
len(Indexes)
```

```python
# 学習データの作成
feature_sup = np.array([[0 for i in xrange(len(supvis))] for j in xrange(326)])
feature_sto = np.array([[0 for i in xrange(len(story))] for j in xrange(326)])
feature_com = np.array([[0 for i in xrange(len(company))] for j in xrange(326)])
print feature_sup.shape,feature_sto.shape,feature_com.shape
```

```python
data_feature = np.array(data_feature)

for i in xrange(326):
    for j in xrange(len(supvis)):
        if data_feature[:,1][i]==supvis[j]:
            feature_sup[i][j]=1
            #break
    for j in xrange(len(story)):
        if data_feature[:,2][i]==story[j]:
            feature_sto[i][j]=1
            #break
    for j in xrange(len(company)):
        if data_feature[:,3][i]==company[j]:
            feature_com[i][j]=1
            #break
```

```python
# 学習データ完成
feature_learn = np.concatenate([np.concatenate([feature_sup,feature_sto],axis=1),feature_com],axis=1)
```

```python
#　学習データの書き出し
np.savetxt('feature_learn.csv',feature_learn,delimiter=',',fmt='%d')
```

```python
# テストデータの作成
feature_supt = np.array([[0 for i in xrange(len(supvis))] for j in xrange(50)])
feature_stot = np.array([[0 for i in xrange(len(story))] for j in xrange(50)])
feature_comt = np.array([[0 for i in xrange(len(company))] for j in xrange(50)])

data_feature2017 = np.array(data_feature2017)

for i in xrange(50):
    for j in xrange(len(supvis)):
        if data_feature2017[:,1][i]==supvis[j]:
            feature_supt[i][j]=1
            #break
    for j in xrange(len(story)):
        if data_feature2017[:,2][i]==story[j]:
            feature_stot[i][j]=1
            #break
    for j in xrange(len(company)):
        if data_feature2017[:,3][i]==company[j]:
            feature_comt[i][j]=1
            #break

feature_test = np.concatenate([np.concatenate([feature_supt,feature_stot],axis=1),feature_comt],axis=1)
```

```python
feature_test
```

```python
#　テストデータの書き出し
np.savetxt('feature_test.csv',feature_test,delimiter=',',fmt='%d')
```
